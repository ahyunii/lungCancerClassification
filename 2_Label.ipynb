```python
import pandas as pd
import os
```

```python
#load csv file and select columns
org=pd.read_csv('D:/NSCLC Radiomics Lung1.clinical-version3-Oct 2019.csv',sep=',')
org=org[['PatientID','Survival.time']]
org
```

```python
#our patient list to DataFrame
fileID = os.listdir('D:/TCIA_real/patches')
patID = []
for x in fileID:
    patID.append('LUNG1-'+x[:3])
new=pd.DataFrame(patID,columns=['PatientID'])
new
```

```python
#merge org with new , key='PatientID'
#new=pd.merge(org,new,on='PatientID')
new=pd.merge(org,new)
new
```

```python
#convert to month unit and check for recurrence
survival_month=[]
recurrence=[]
for i in range(len(new)):
    survival_month.append(round(new['Survival.time'][i]/30))
    if survival_month[i]<=24:
        recurrence.append(1)
    else:
        recurrence.append(0)
```

```python
#add new columns
new['Survival.time(month)']=survival_month
new['recurrence']=recurrence
```
